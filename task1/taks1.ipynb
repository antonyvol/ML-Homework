```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def 
# sgd
def sgd(model, x, y, lr):
    
    return model.coef + lr * (y - model(x)*x)
```

```python
n = 100

x = np.random.uniform(0.0, 1.0, n)
e = np.random.uniform(-0.3, 0.3, n)
y = np.sin(2*np.pi*x) + e
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
```

```python
poly_order = 5
coef = np.random.uniform(-0.5, 0.5, poly_order)
model = np.polynomial.polynomial.Polynomial(coef)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.scatter(x, model(x))
```

```python
model.coef
```

```python
lr = 0.1
for _ in range(100):
    t = sgd(model, x, y, lr)
    model.coef = t
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.scatter(x, model(x))
```
