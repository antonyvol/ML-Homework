```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def poly(c, x):
    return c[0] + c[1]*x + c[2]*(x**2) + c[3]*(x**3) + c[4]*(x**4)
    

# sgd DOES NOT WORK
def sgd(c, x, y, lr):
    nc = c
    m = len(x)
    for i in range(m):
        for j in range(5):
            nc[j] = c[j] + lr * (y[i] - poly(c, x[i])) # * x[i]
    return nc
```

```python
n = 100

x = np.random.uniform(0.0, 1.0, n)
e = np.random.uniform(-0.3, 0.3, n)
y = np.sin(2*np.pi*x) + e
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
```

```python
c = np.random.uniform(-0.5, 0.5, 5)
y_pred = np.array([poly(c, v) for v in x])
```
