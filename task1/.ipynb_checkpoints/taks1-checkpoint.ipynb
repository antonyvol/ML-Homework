```python
import math
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python

def sgd(c, x, y, lr):
    ypred = np.poly1d(c)(x) 
    return np.array([c[j] + lr * ((y-ypred) * (x**(len(c)-j-1))).sum() for j in range(len(c))])
```

```python
n = 100

x = np.random.uniform(0.0, 1.0, n)
e = np.random.uniform(-0.3, 0.3, n)
y = np.sin(2*np.pi*x) + e
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
```

```python
c = np.random.uniform(-0.5, 0.5, 4)
y_pred = np.poly1d(c)(x)
```

```python
y_pred
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.scatter(x, y_pred)
```

```python
lr = 0.01
nc = copy.copy(c)
error = []
for _ in range(20000):
    nc = sgd(nc, x, y, lr)
    error.append(np.linalg.norm(y-np.poly1d(nc)(x)))
y_pred = np.poly1d(nc)(x)
```

```python
y_pred
```

```python
#Final Model 
print(np.poly1d(nc))
```

```python
plt.figure(figsize=(15,7))
plt.xlabel("Data Points")
plt.ylabel("Output")
plt.scatter(x, y, label="Y")
plt.scatter(x, y_pred, label="Y_pred")
plt.legend(prop={'size': 20})
```

```python
plt.figure(figsize=(15,7))
plt.scatter(range(20000), error)
plt.xlabel("Iterations")
plt.ylabel("Error")
```
